```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import os
```

```python
path = "C:\\Users\\asiak\\OneDrive\\Pulpit\\pogoda\\output\\output_csv15\\"
```

```python
files = [f for f in os.listdir(path)]
files.pop(-1)
print(files)
```

```python
all_data = []
for file in files:
    filepath = path + file
    tmp = pd.read_csv(filepath)
    all_data.append(tmp)
df = pd.concat(all_data)
```

```python
df
```

```python
df['country'].value_counts()
```

```python
filtered_df = df.groupby('country').filter(lambda x: len(x) > 1000)

# Calculate mean temperature for each country and reset index for Seaborn
mean_temperatures = filtered_df.groupby('country')['mean_temp'].mean().reset_index()

# Sort the DataFrame in ascending order of mean temperature and select top 15
sorted_df = mean_temperatures.sort_values(by='mean_temp')
top_15_coldest = sorted_df.head(15)
top_15_hottest = sorted_df.tail(15)

# Plotting the top 15 coldest countries
plt.figure(figsize=(10, 6))
sns.barplot(x='mean_temp', y='country', data=top_15_coldest, palette="cool")
plt.xlabel('Mean Temperature')
plt.ylabel('Country')
plt.title('Top 15 Coldest Countries')
plt.show()

# Plotting the top 15 hottest countries
plt.figure(figsize=(10, 6))
sns.barplot(x='mean_temp', y='country', data=top_15_hottest, palette="hot")
plt.xlabel('Mean Temperature')
plt.ylabel('Country')
plt.title('Top 15 Hottest Countries')
plt.show()
```

```python
filtered_df = df.groupby('city').filter(lambda x: len(x) > 3)

# Calculate mean temperature for each country and reset index for Seaborn
mean_temperatures = filtered_df.groupby('city')['mean_temp'].mean().reset_index()

# Sort the DataFrame in ascending order of mean temperature and select top 15
sorted_df = mean_temperatures.sort_values(by='mean_temp')
top_15_coldest = sorted_df.head(15)
mod_cold = top_15_coldest.merge(df, on='city', how='left')[['city', 'country']].drop_duplicates()
mod_cold['city1'] = mod_cold.apply(lambda x: x.city+' ('+x.country+')', axis=1)
top_15_coldest = top_15_coldest.merge(mod_cold, on='city', how='left')
top_15_coldest = top_15_coldest[['city1','mean_temp']]
top_15_coldest = top_15_coldest.rename(columns={'city1':'city'})
top_15_hottest = sorted_df.tail(15)
mod_hot = top_15_hottest.merge(df, on='city', how='left')[['city', 'country']].drop_duplicates()
mod_hot['city1'] = mod_hot.apply(lambda x: x.city+' ('+x.country+')', axis=1)
top_15_hottest = top_15_hottest.merge(mod_hot, on='city', how='left')
top_15_hottest = top_15_hottest[['city1','mean_temp']]
top_15_hottest = top_15_hottest.rename(columns={'city1':'city'})

# Plotting the top 15 coldest countries
plt.figure(figsize=(10, 6))
sns.barplot(x='mean_temp', y='city', data=top_15_coldest, palette="cool")
plt.xlabel('Mean Temperature')
plt.ylabel('city')
plt.title('Top 15 Coldest cities')
plt.show()

# Plotting the top 15 hottest countries
plt.figure(figsize=(10, 6))
sns.barplot(x='mean_temp', y='city', data=top_15_hottest, palette="hot")
plt.xlabel('Mean Temperature')
plt.ylabel('city')
plt.title('Top 15 Hottest cities')
plt.show()
```
